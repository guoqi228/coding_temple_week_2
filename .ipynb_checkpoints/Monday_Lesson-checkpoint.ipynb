## Exercise #1 <br>
<p>Build a pyramid of x's with 5 rows...</p>

```python
n = 5 # number of rows
def x_pyramid(n):
    num_slot = n*2 - 1 
    for i in range(n):
        x_list = [' '] * num_slot
        start = num_slot//2 - i
        end = start + i*2 + 1
        for j in range(start, end, 2):
            x_list[j] = 'X'
        x_string = ''.join(x_list)
        print(x_string)
        
x_pyramid(20)
```

```python
def plot_pyramid(n):
    for row in range(n):
        flag = True
        for col in range(row):
            if flag == True:
                print('O'*(n - row), end = ' ')
                flag =False
            print('X', end = " ")
        print('O')

plot_pyramid(5)
```

## Exercise #2 <br>
<p>Cube Number Test... Find all perfectly cubed numbers up to 1000</p>

```python
for num in range(2,100):
    if num**3 < 100:
        print(num ** 3)
```

```python
def give_cube(n):
    
    num = 0
    num_list = []
    while True:
        cube = num**3
        if cube <= n:
            num_list.append(cube)
            num += 1
        if cube > n:
            break
    return num_list

give_cube(100)
    
```

## Exercise #3 <br>
<p>Get first 100 Prime Numbers</p>

```python
for num in range(100):
    if num > 1:
        for index in range(2, num):
            if num % index == 0:
                break
        else:
            print(num, end = " ")
            
```

```python
def is_prime(n):
    if n == 1 or n == 0: 
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False
        elif n % i != 0:
            continue
    return True

def give_n_prime(n):
    start = 2
    a_list = []
    counter = 0
    while True:
        if counter > n:
            break
        if is_prime(start):
            a_list.append(start)
            counter += 1
            start += 1
        else:
            counter += 1
            start += 1
    print(a_list)

give_n_prime(100)
    
```

## Exercise #4 <br>
<p>Make an addition calculator that takes in two inputs from the user and performs addition on those two numbers...<br>Then check to make sure that both numbers are instances of int or float<br><i>Hint: Use int() and float() to convert string inputs to numbers</i></p>

```python
def add_two_nums():
    while True:
        
        is_num1_number = False
        while not is_num1_number:
            try:
                num1 = float(input('Enter first number: '))
                is_num1_number = True
            except:
                print('Please enter a number!')
                
        is_num2_number = False
        while not is_num2_number:
            try:
                num2 = float(input('Enter second number: '))
                is_num2_number = True
            except:
                print('Please enter a number!')
        
        ans = num1 + num2
        print('Sum of {} and {} is: {}'.format(num1, num2, ans))
        
        is_n = False
        while not is_n:
            try:
                keep_going = str(input('Do you want to continue? y or n:'))
                if keep_going.lower() == 'n':
                    is_n = True
                elif keep_going.lower() == 'y':
                    break
            except:
                print('Please enter y or n!')
                
        if keep_going.lower() == 'n':
            print('Bye!')
            break
        else:
            continue
            
add_two_nums()
        
```

## Exercise #5 <br>
<p><b>Extra Credit</b><br>Do the same as exercise 4, except ask the user what they would like to do (add, subtract, multiply, or divide) then perform that operation...</p>

```python
try:
    num1 = float(input('Enter number 1: '))
    num2 = float(input('Enter number 2: '))
    operation = str(input("Enter 'add', 'substract', 'multiply' or 'divide':"))
    if operation.lower() == 'add':
        print('{} + {} = {}'.format(num1, num2, num1+num2))
    elif operation.lower() == 'substract':
        print('{} - {} = {}'.format(num1, num2, num1-num2))
    elif operation.lower() == 'multiply':
        print('{} * {} = {}'.format(num1, num2, num1*num2))
    elif operation.lower() == 'divide':
        print('{} / {} = {}'.format(num1, num2, num1/num2))
    else:
        print('Operation not recognized!')
except:
    print('Something went wrong...try again!')
```
