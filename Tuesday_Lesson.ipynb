### Remove duplicate in a list

```python
a_list = ['conner', 'conner', 'conner','evan','max','evan']

def remove_dup(a_list):
    a_list = sorted(a_list)
    out = [a_list[0]]
    n = len(a_list)
    for i in range(1,n-1):
        if a_list[i] != a_list[i + 1]:
            out.append(a_list[i + 1])
    return out

remove_dup(a_list)           
```

```python
def remove_dup_2(a_list):
    a_list = sorted(a_list)
    n = len(a_list)
    out = [a_list[0]]
    left = 0
    right = 1
    while right < n:
        if a_list[left] != a_list[right]:
            out.append(a_list[right])
            left = right
        right += 1
    return out

remove_dup_2(a_list)        
```

### Function exercise
#### write a function that loops through a list of first name and last name, combine this two into full name.

```python
first_name = ['Connor', 'Evan', 'Jordan', 'Max', 'kj;kl']
last_name = ['Milliken', 'Smith', 'Williams', 'Bell']

def full_name(first_list, last_list):
    '''
        param:  two lists - list of first name and list of last name
        return: a list of full name
    '''
    len_first = len(first_list)
    len_last = len(last_list)
    
    if len_first < len_last:
        for i in range(len_last - len_first):
            first_list.append('______')
    else:
        for i in range(len_first - len_last):
            last_list.append('______')
    
    out = []
    for i in range(len(first_list)):
        out.append(first_list[i] + ' ' + last_list[i])
    return out

full_name(first_name, last_name)
```

```python
first_name = ['Connor', 'Evan', 'Jordan', 'Max']
last_name = ['Milliken', 'Smith', 'Williams', 'Bell']

def full_name_2(first_list, last_list):
    '''
        param:  two lists - list of first name and list of last name
        return: a list of full name
    '''
    return [name1 + ' ' + name2 for (name1, name2) in zip(first_name, last_name)] 

full_name_2(first_name, last_name)
```

## Exercise 1 <br>
<p>Write a function that performs a bubble sort<br><i><b>Hint: in place assignment is 'num[x], num[x+1] = num[x+1], num[x]'</b></i></p>

```python
#nums = [10,8,9,6,7,8,6,5,4,3,2]
nums = [10,9,8,7,7,7,7,6,5,4,3,2,1,0]

def bubble_sort(nums):
    
    n = len(nums)
    
    for j in range(n-1):
        for i in range(n-1):
            if nums[i] > nums[i + 1]:
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
            elif nums[i] < nums[i + 1]:
                continue
                
    return nums

bubble_sort(nums)
        
```

```python
nums = [10,9,8,7,7,7,7,6,5,4,3,4,5,2,1,0]



def bubble_sort_2(nums):
    '''
    param: a list a number
    return: a sorted list (ascending)
    '''
    n = len(nums)
    while True:
        swap = False
        for i in range(n-1):
            if nums[i] > nums[i + 1]:
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                swap = True
            elif nums[i] < nums[i + 1]:
                continue
        if swap == False:
            return nums
            break

bubble_sort(nums)
```

## Exercise 3 <br>
<p> Ask the user to input a number and check if that number is prime using a function, loop until they no longer want to check a number.<br><b><i>Hint: Use yesterday's prime number code</i></b></p>

```python
def check_prime(num):
    if num == 0 or num == 1:
        return False
    for i in range(2,num):
        if num%i == 0:
            return False
        else:
            continue
    return True

def keep_checking():
    keep_going = True
    while keep_going:
        
        # keep asking for a number until input is a number
        is_num = False
        while not is_num:
            try:
                num = int(input('Input a number to check if it\'s prime: '))
                is_num = True
                if is_num:
                    if check_prime(num):    
                        print('{} is a prime number.'.format(num))
                    else:
                        print('{} is not a prime number.'.format(num))
            except:
                print('Please re-enter, make sure it\'s a integer number!')
                
        # keep asking for input until get a 'n' or 'y'
        is_n_or_y = False    
        while not is_n_or_y:
            try:
                is_n_or_y = str(input('Do you want to continue? Enter y or n: '))
                if is_n_or_y.lower() == 'n':
                    is_n_or_y = True
                    print('Bye!')
                    keep_going = False
                elif is_n_or_y.lower() == 'y':
                    is_n_or_y = True
                else:
                    print('Please enter y or n!')
            except:
                print('Please enter y or n!')

keep_checking()   
```

## Exercise 4 <br>
<p>Ask the user to input a number of rows they'd like to build a pyramid to... then build a pyramid to that specific number of rows.<br><b><i>Hint: Use the pyramid code from yesterday and throw it into a function</i></b></p>

```python
def x_pyramid(n):
    num_slot = n*2 - 1 
    for i in range(n):
        x_list = [' '] * num_slot
        start = num_slot//2 - i
        end = start + i*2 + 1
        for j in range(start, end, 2):
            x_list[j] = 'X'
        x_string = ''.join(x_list)
        print(x_string)
        
x_pyramid(20)
```
