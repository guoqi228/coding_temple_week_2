## Class Methods <br>
<p>While inside of a class, functions are referred to as 'methods'. If you hear someone mention methods, they're speaking about classes. Methods are essentially functions, but only callable on the instances of a class.</p>

##### Creating

```python
class Car():
    
    def __init__(self, wheels, color, make, model, odometer):
        self.wheels = wheels
        self.color = color
        self.make = make
        self.model = model
        self.odometer = odometer
        
    def print_class():
        print('Information')
        
    def print_info(self):
        print(f'Wheels: {self.wheels}')
        print(f'Color: {self.color}')
        print(f'Make: {self.make}')
        print(f'Model: {self.model}')
        print(f'Odometer: {self.odometer}')
        print('\n')
        
    def update_color(self, new_color):
        self.color = new_color
        
    def increment_odometer(self):
        self.odometer += 199
        
        
Car.print_class()
car_1 = Car(4, 'Red', 'Ford', 'Escape',9000)
car_1.print_info()

car_2 = Car(color="Blue", make='Honda',wheels=5, model='Civic', odometer = 8900)
car_2.print_info()

car_2.update_color('White')
car_2.print_info()

car_2.increment_odometer()
car_2.print_info()
```

## Inheritance <br>
<p>You can create a child-parent relationship between two classes by using inheritance. What this allows you to do is have overriding methods, but also inherit traits from the parent class. Think of it as an actual parent and child, the child will inherit the parent's genes, as will the classes in OOP</p>

##### Syntax for Inheriting from a Parent Class

```python
class Animal():
    def __init__(self, speed, legs = 4):
        self.legs = legs
    
    def make_sound(self):
        print('Animal sound')
        
    def get_speed(self):
        return self.speed
    
class Dog(Animal):
    def __init__(self, speed, legs, name):
        super().__init__(speed, legs)
        self.name = name
        
dog_1 = Dog(14, 4, 'Sammi')
dog_1.name
        
```

##### In-Class Exercise #4 - Create a class 'Ford' that inherits from 'Car' class and initialize it as a Blue Ford Explorer with 4 wheels using the super() method

```python
class Car():
    
    def __init__(self, year, color, doors, seats):
        self.year = year
        self.color = color
        self.doors = doors
        self.seats = seats
        
    def print_info(self):
        print('Year: {}'.format(self.year))
        print('Color: {}'.format(self.color))
        print('Doors: {}'.format(self.doors))
        print('Seats: {}'.format(self.seats))
        print('\n')
        
        
class Ford(Car):
      
    def __init__(self, year, color, doors, seats, model, make = 'Ford'):
        super().__init__(year, color, doors, seats)
        self.make = make
        self.model = model
        
    def print_info(self):
        print('Year: {}'.format(self.year))
        print('Make: {}'.format(self.make))
        print('Model: {}'.format(self.model))
        print('Color: {}'.format(self.color))
        print('Doors: {}'.format(self.doors))
        print('Seats: {}'.format(self.seats))
        print('\n')

car_1 = Car(color='Blue', doors= 5, seats= 7, year= '2013')
car_1.print_info()
        
ford_1 = Ford(year= '2011', color= 'Red', model='Escape', doors= 4, seats=4)
ford_1.print_info()
       
```

### Exercise 1 - Turn the shopping cart program from yesterday into an object-oriented program

```python
class Menu(): 
    
    def __init__(self, menu = ['Show all items', 'Add item to cart', 'Remove item from cart', 'Show items in cart',
                              'Check out', 'Cancel']):
        self.menu = menu
        
    def show_menu(self):
        print('Menu')
        print('========================================')
        for i, item in enumerate(self.menu):
            print('{} -----> {}'.format(i, item))
        print('========================================')
    
    def menu_length(self):
        return len(self.menu)
    
    def modify_menu(self):
        print('You can modify the menu now.')
        self.show_menu()
        is_num = False
        while not is_num:
            try:
                option = int(input('Please enter the option you want to modify: '))
                if option not in range(self.menu_length()):
                    print('Please enter a valid number: 0 to {}!'.format(self.menu_length() - 1))
                else:
                    new_name = input('Please modify option {}: '.format(option))
                    self.menu[option] = new_name
                    print('Option {} is modified to: {}'.format(option, new_name))
                    is_num = True
            except:
                print('Please enter a valid number: 0 to {}!'.format(self.menu_length() - 1))
                        
```

```python
class AllItems():
    
    def __init__(self, items = [('Chip', 3.98), ('Beer', 6.99), ('Soda', 5.98), 
                                ('Chicken',11.99), ('Beef',15.99), ('Chocolate',5.02), 
                                ('Juice',6.98), ('Wine',15.98)]):
        self.items = items
        
    def show_all_items(self):
        print('All items')
        print('========================================')
        for i, item in enumerate(self.items):
            print('{}: {}......${}'.format(i, item[0], item[1]))
        print('========================================') 
        
    def num_items(self):
        return len(self.items)
    
    def choose_item(self):
        is_num = False
        while not is_num:
            try:
                num = int(input('Choose a item: '))
                if num not in range(self.num_items()):
                    print('Please enter a valid item number from 0 to {}!'.format(self.num_items() - 1))
                else:
                    return self.items[num]
                    print('Item {} chosen successful!'.format(num))
                    print('========================================')
                    is_num = True
            except:
                print('Please enter a valid item number!')          
    
    def add_item(self):
        self.show_all_items()
        print('You can now add a new item.')
        
        is_string = False
        while not is_string:
            name = str(input('Enter the name of the item: '))
            if name.isalpha():
                is_string = True
            else:
                print('Item name can only be string!')
                continue
        
        is_num = False
        while not is_num:
            try:
                price = float(input('Enter the price of the item:'))
                if price >= 0:
                    self.items.append((name, price))
                    print('Item add successful!')
                    print('========================================')
                    is_num = True
                else:
                    print('Please enter a valid price!')
                    continue
            except:
                print('Please enter a valid price!')
    
    def delete_item(self):
        self.show_all_items()
        print('You can now delete a item.')
        
        is_num = False
        while not is_num:
            try:
                num = int(input('Enter the item you want to delete: '))
                if num not in range(self.num_items()):
                    print('Please enter a valid item number from 0 to {}!'.format(self.num_items() - 1))
                else:
                    del self.items[num]
                    print('Item delete successful!')
                    print('========================================')
                    is_num = True
            except:
                print('Please enter a valid item number!')  
                
    def update_price(self):
        self.show_all_items()
        print('You can now update the price of a item.')
        
        is_num = False
        while not is_num:
            try:
                num = int(input('Enter the item you want to update: '))
                if num not in range(self.num_items()):
                    print('Please enter a valid item number from 0 to {}!'.format(self.num_items() - 1))
                else:
                    name = self.items[num][0]
                    del self.items[num]
                    is_num = True
                    
                    is_price_valid = False
                    while not is_price_valid:
                        try:
                            price = float(input('Enter the price to update: '))
                            if price >= 0:
                                self.items.insert(num, (name, price))
                                print('Item update successful!')
                                print('========================================')
                                is_price_valid = True
                            else:
                                print('Please enter a valid price!')
                                continue
                        except:
                            print('Please enter a valid price!')
                            
            except:
                print('Please enter a valid item number!')  
                               
```

```python
class ShoppingCart():
    
    def __init__(self, cart_items = []):
        #if cart_items == None:
        #    cart_items = []
        self.cart_items = cart_items
    
    def num_cart_items(self):
        return len(self.cart_items)
    
    def show_cart(self):
        if self.num_cart_items() == 0:
            print('Shopping cart is empty!')
            print('========================================')

        else:
            print('All items in shopping cart')
            print('========================================')
            for i, item in enumerate(self.cart_items):
                print('{}: {}......${}'.format(i, item[0], item[1]))
            print('========================================')

    def add_to_cart(self, item):
        self.cart_items.append(item)
        print('Iterm added to cart!')
        print('========================================')
                  
    def remove_from_cart(self):
        if self.num_cart_items() == 0:
            print('Shopping cart is empty!')
            print('========================================')
            return None
        else:
            self.show_cart()
                  
            is_num = False
            while not is_num:
                try:
                    num = int(input('Enter the item you want to remove: '))
                    if num not in range(self.num_cart_items()):
                        print('Please enter a valid item number from 0 to {}!'.format(self.num_cart_items() - 1))
                    else:
                        del self.cart_items[num]
                        print('Item remove successful!')
                        print('========================================')
                        is_num = True
                except:
                    print('Please enter a valid item number!') 
    
    def check_out(self):
                  
        if self.num_cart_items() == 0:
            print('Your cart is empty!')
            print('========================================')
            return False
        print('Items in your shopping cart: ')
        self.show_cart()
        print('Checking out...')
        print('========================================')
        total_price = 0
        for item in self.cart_items:
            total_price += item[1]
        print('Your total is: ${}'.format(total_price))
        print('Thanks for shopping with us.') 
        return True
                  
    def cancel(self):
        print('Transation cancelled!')
        return True
```
